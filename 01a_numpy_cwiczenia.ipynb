```python
def almost_variance(x: np.ndarray) -> np.ndarray:
    """Oblicza 1/n * SUM (x_i - mean(x))^4"""
    m = sum(x) / len(x)
    result = 0
    for i in range(len(x)):
        result += (x[i] - m) ** 4
    result /= len(x)
    return result


def numpy_almost_var(x: np.ndarray) -> np.ndarray:
    m = np.mean(x)
    powers = (x - m) ** 4
    return np.mean(powers)


X = np.random.random(10000)
```

```python
%%timeit
almost_variance(X)
```

```python
%%timeit
numpy_almost_var(X)
```

```python
def poly(x: int, a: np.ndarray) -> np.ndarray:
    # Tworzymy tablicę potęg x: [x^0, x^1, x^2, ..., x^k]
    powers = np.arange(len(a))
    
    # Mnożymy współczynniki a przez odpowiednie potęgi x
    result = a * (x ** powers)
    
    # Sumujemy wyniki, aby uzyskać wartość wielomianu
    return np.sum(result)

print("Wynik funkcji poly:", poly(3, np.array([1, 2, 4])))
```

```python
X_diag_normal = np.random.multivariate_normal([-5, 2.0], [[10, 0], [0, 0.5]], size=200)
X_multivariate_normal = np.random.multivariate_normal([-5, -2], [[1, 0.99], [0.99, 1]], size=500)
X_uniform = np.random.normal(6, 5, size=(50, 2))
X_gmm = np.concatenate((X_multivariate_normal, X_diag_normal), axis=0)

utils.scatter_with_whiten(X_diag_normal, whiten, name="Independent normal distribution")
utils.scatter_with_whiten(X_multivariate_normal, whiten, name="Multivariate normal distribution")
utils.scatter_with_whiten(X_uniform, whiten, name="Uniform distribution")
utils.scatter_with_whiten(X_gmm, whiten, name="Gaussian mixture distribution")
```

```python
from scipy.linalg import sqrtm

def whiten(X: np.ndarray) -> np.ndarray:
    # 1. Obliczamy średnią dla każdego wymiaru
    meanX = np.mean(X, axis=0)
    
    # 2. Odejmujemy średnią od danych
    X_centered = X - meanX
    
    # 3. Transponujemy macierz i obliczamy macierz kowariancji
    covX = np.cov(X_centered, rowvar=False)  # `rowvar=False`, ponieważ kolumny są cechami
    
    # 4. Obliczamy odwrotność pierwiastka macierzy kowariancji
    covX_inv_sqrt = np.linalg.inv(sqrtm(covX))
    
    # 5. Zastosowanie whitening: (X - meanX) * (covX)^(-1/2)
    X_whitened = X_centered.dot(covX_inv_sqrt)
    
    return X_whitened

# Przykład użycia funkcji
X = np.random.multivariate_normal([5, -2], [[10, 3], [3, 2]], size=500)
X_whitened = whiten(X)
print(X_whitened)
```

## Standaryzacja a whitening
W pierwszym notebooku implementowaliśmy następującą funkcję, która wykonuje zadanie podobne do whiteningu:

$$ f(x_{ij}) = \frac{x_{ij} - \mu_j}{\sigma_j} $$

Tę funkcję nazwijmy **standaryzacją**. Czym zatem różni się standaryzacja od whiteningu? Czy nie wystarczy znormalizować kolumn i zamiast tego musimy wykonywać znacznie bardziej kosztowną operację liczenia i odwracania macierzy kowariancji?

```python
utils.scatter_with_whiten(
    X_diag_normal, whiten, name="Independent normal distribution", standarize=True
)
utils.scatter_with_whiten(
    X_multivariate_normal, whiten, name="Multivariate normal distribution", standarize=True
)
utils.scatter_with_whiten(X_uniform, whiten, name="Uniform distribution", standarize=True)
utils.scatter_with_whiten(X_gmm, whiten, name="Gaussian mixture distribution", standarize=True)
```

```python
import numpy as np
from scipy.linalg import sqrtm

# Standaryzacja: f(xij) = (xij - uj) / rj
def standardize(X: np.ndarray) -> np.ndarray:
    # Oblicz średnią i odchylenie standardowe dla każdej kolumny
    mean = np.mean(X, axis=0)
    std_dev = np.std(X, axis=0)
    
    # Standaryzuj: (X - mean) / std_dev
    return (X - mean) / std_dev

# Whitening z regularizacją
def whiten(X: np.ndarray, epsilon: float = 1e-5) -> np.ndarray:
    # 1. Odjęcie średniej od każdej cechy
    X_centered = X - np.mean(X, axis=0)
    
    # 2. Obliczenie macierzy kowariancji
    cov = np.cov(X_centered, rowvar=False)
    
    # 3. Dodanie małej wartości do diagonalnych elementów macierzy kowariancji (regularizacja)
    cov += epsilon * np.eye(cov.shape[0])
    
    # 4. Obliczenie odwrotności pierwiastka macierzy kowariancji
    cov_inv_sqrt = np.linalg.inv(sqrtm(cov))
    
    # 5. Przekształcenie macierzy X na postać whitened
    X_whitened = X_centered @ cov_inv_sqrt
    return X_whitened

# Przykładowe dane
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])

# Zastosowanie standaryzacji
X_standardized = standardize(X)
print("Standaryzacja:\n", X_standardized)

# Zastosowanie whiteningu z regularizacją
X_whitened = whiten(X)
print("Whitening:\n", X_whitened)
```
